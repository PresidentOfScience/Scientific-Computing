```python
import numpy as np

# Validasi apakah dia diagonally dominant

def gaussSeidel():
  # Buat dua array penampung [0, 0, 0]
  x = np.zeros(3)
  x_old = np.zeros(3)
  target = 1E-5
  err = 1
  while(err > target):
    # Kalkulasi per iterasi
    x[0] = (19 - x[1] - 2*x[2])/3
    x[1] = (20 - 2*x[0] - x[2])/5
    x[2] = (17 - x[0] - x[1])/3
    # Bandingkan nilai sekarang dan sebelumnya
    err = np.abs(np.sum(x - x_old))
    print(f"x = {x[0]}, y = {x[1]}, z = {x[2]}")
    # Masukkan nilai sekarang ke nilai lama
    x_old = np.array(x)

myArray = np.array([
    [3, 1, 2],
    [2, 5, 1],
    [1, 1, 3]
])

# Jadikan absolut
myArray = np.abs(myArray)

# Cari diagonal
myDiag = np.diag(myArray)

# Cari nondiagonal
nonDiag = np.sum(myArray, axis=1) - np.diag(myArray)

if(np.all(myDiag < nonDiag)):
  print("Diagonally non-dominant")
else:
  print("Diagonally dominant")
  gaussSeidel()

```

```python
var = np.array([
    [3, 1, 2],
    [2, 5, 1],
    [1, 1, 3]
])

konstanta = np.array([19, 20, 17])

print(np.linalg.solve(var, konstanta))
```
