```python
import numpy as np
import matplotlib.pyplot as plt

# x^2 - 2

# Function
def equation(number):
    return (1/number) - 0.25

# Lambda
eq = lambda number: (1/number) - 0.25

print("Normal Function : ", equation(2))
print("Normal Function : ", eq(2))

x_new = np.linspace(1,10,1000)
# print(x_new)
plt.plot(x_new, eq(x_new), "r")
```

```python
# 1. Bisection

def bisection(f, a, b, e):
    # Validasi kalau persamaan tersebut memiliki akar
    if(np.sign(f(a)) == np.sign(f(b))):
        return "No root found"
    # Mencari middle (a+b)/2
    middle = (a+b)/2
    # Validasi apakah nilai m sudah memuaskan (epsilon)
    if(np.abs(f(middle)) < e):
        return middle
    # Misal hasilnya negatif, maka pindahkan a ke mid
    if(np.sign(f(middle)) == np.sign(f(a))):
        plt.plot(a, f(a), "bo")
        return bisection(f, middle, b, e)
    # Misal hasilnya positif, maka pindahkan b ke mid
    if(np.sign(f(middle)) == np.sign(f(b))):
        plt.plot(b, f(b), "bo")
        return bisection(f, a, middle, e)

result = bisection(eq, 1, 6, 0.0000000001)
print(result)
```

```python
# 2. Newton Raphson

# Xn+1 = Xn - (f(Xn)/f'(Xn))

f = lambda x: (1/x) - 0.25
dxf = lambda x: -1/(x**2)

# Validasi apakah sudah memuaskan
# Rekursif dan masukkan nilai baru

def NewtonRaphson(f, x, dxf, e):
    if(f(x) < e):
        return x
    else:
        plt.plot(x, f(x), "bo")
        return NewtonRaphson(f, x-(f(x)/dxf(x)), dxf, e)

print(NewtonRaphson(f, 2, dxf, 0.000000001))

plt.plot(x_new, f(x_new), "r-")
```
